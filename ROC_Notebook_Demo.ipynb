```python
%matplotlib inline
import numpy as np
import ROC_demo_v3 as demo3
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets

path = 'ROCs/'
separations = np.array([12, 18, 24])
ld_separations = separations/6
mags = np.arange(12,14.5,0.5)

my_tpfs = np.zeros((len(separations), len(mags), 100))
my_fpfs = np.linspace(0.001,1.0,100)

for i in xrange(len(separations)):
    for j in xrange(len(mags)):
        fpf_init = np.loadtxt(path + str(separations[i])+'_'+str(mags[j])+'_FPF.txt')
        tpf_init = np.loadtxt(path + str(separations[i])+'_'+str(mags[j])+'_TPF.txt')
        my_tpfs[i,j,:] = np.interp(my_fpfs, fpf_init, tpf_init)
            
            
interact(demo3.display, val=widgets.FloatSlider(min=0.001,max=1.0,step=0.01,value=0.1), my_tpfs=fixed(my_tpfs), my_fpfs=fixed(my_fpfs), mags=fixed(mags), ld_separations=fixed(ld_separations))
#demo.display()
```
